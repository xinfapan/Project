```python
store=pd.read_csv('Superstore.csv') 
store.describe()
```

```python
pd.options.display.max_columns = None  #set to see all column
store.head(5)
```

```python
#check null

print(store.isnull().sum())

#check duplicate
print('check duplicate')
duplicate = store[store.duplicated()]
 
print(duplicate)
 
#no duplicate
```

```python
#sales
fig=plt.figure()
ax = fig.add_subplot(1,1,1)
#Variable
ax.boxplot(store['Sales'])
plt.show()
#it seems like majority of sales in under 5000
```

```python
f = plt.figure()
f.set_figwidth(20)
plt.scatter(store['Sub-Category'],store.Sales)
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.set_xlabel('subcategory')
ax1.set_ylabel('Sum of Sales')
ax1.set_title("subcategory wise Sum of Sales")

sales_sub = store.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False)
#groupby state and sum of sales

sales_sub.plot(kind='bar')
```

```python


fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.set_xlabel('States')
ax1.set_ylabel('Sum of Sales')
ax1.set_title("State wise Sum of Sales")

sales_State = store.groupby("State")["Sales"].sum().sort_values(ascending=False)
#groupby state and sum of sales

sales_State.plot(kind='bar')
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.set_xlabel('States')
ax1.set_ylabel('count of order')
ax1.set_title("State wise count of order")

sales_order = store.groupby("State")["Order ID"].count().sort_values(ascending=False)
#that is why california has most sales, bc more order

sales_order.plot(kind='bar')
```

```python

print(store.columns) #variable error, lets see what the column names are

#lets say I want top 5 sales state only
top5=["California", "New York", "Texas", "Washigton","Pennsyvania"]
sale_top5 = store[store["State"].isin(top5)]


sales_cat = sale_top5.groupby(["State","Category"])["Sales"].sum().sort_values(ascending=False) 
sales_cat.head()


fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.set_xlabel('States')
ax1.set_ylabel('sum of sales')
ax1.set_title("State wise category sales")
sales_cat.plot(kind='bar')
```
